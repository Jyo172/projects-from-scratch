```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data1 =np.random.normal(0,0.1,1000)
data2 =np.random.normal(1,0.4,1000)+np.linspace(0,1,1000)
data3 =2+np.random.random(1000)*np.linspace(1,5,1000)
data4 =np.random.normal(3,0.2,1000)+0.*np.sin(np.linspace(0,20,1000))
```

```python
data = np.vstack([data1,data2,data3,data4]).transpose()
```

```python
df=pd.DataFrame(data,columns=['data1','data2','data3','data4'])
df.head()
```

```python
df.plot(title='Line plot')
```

```python
plt.plot(df)
plt.title('Line plot')
plt.legend(['data1','data2','data3','data4'])
```

```python
df.plot(style='.')
```

```python
_ =  df.plot(kind = 'scatter', x= 'data1', y= 'data2')
```

```python
df.plot(kind='hist',bins=50,title='histogram',alpha=0.4,figsize=(12,8))#alpha to control transparency
```

```python
#cummulative distribution
df.plot(kind='hist',bins=50,title='cummulative histogram',normed=True,cumulative=True,alpha=0.4,figsize=(12,8))
```

```python
#Box Plot
df.plot(kind='box',title='Boxplot')
```

```python
#Subplots
fig,ax= plt.subplots(2,2,figsize=(16,12))

df.plot(ax=ax[0][0],title='Line plot')

df.plot(ax=ax[0][1],style='.',title='Scatter plot')

df.plot(ax=ax[1][0], kind='hist', bins=50, title='Histogram',alpha = 0.4)

df.plot(ax=ax[1][1], kind='box', title='Box plot')

plt.tight_layout()#spacing between plots
```

```python
#Pie_charts
gt01 = df['data1'] >0.1
piecounts = gt01.value_counts()
piecounts
```

```python
piecounts.plot(kind='pie', figsize=(5,5), explode=[0,0.15], labels=['<= 0.1', '> 0.1'], autopct='%1.1f%%', shadow=True, startangle=90, fontsize=16)
```

```python
#Hexbin plot
data = np.vstack([np.random.normal((0,0),2, size=(1000,2)), np.random.normal((9,9), 3, size=(2000,2))])
df=pd.DataFrame(data, columns=['x','y'])
```

```python
df.head()
```

```python
df.plot()
```

```python
df.plot(kind='hexbin', x='x', y='y', bins=100, cmap='rainbow')
```

```python
plt.hexbin(df.x, df.y, gridsize=(15,15) )
plt.show()
```
